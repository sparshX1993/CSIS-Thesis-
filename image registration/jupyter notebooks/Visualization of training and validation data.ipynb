```python
import os
import h5py
import random
import matplotlib.pyplot as plt

def visualize_random_frame(log_folder):
    # Get a list of all HDF5 files in the folder
    log_files = [f for f in os.listdir(log_folder) if f.endswith(".hdf5")]

    if not log_files:
        print("No HDF5 files found in the folder.")
        return

    for log_file in log_files:
        log_path = os.path.join(log_folder, log_file)

        # Extract the log name (excluding the file extension)
        log_name = os.path.splitext(log_file)[0]
        log_name1 = log_name+".hdf5"
        
        main, noise = log_name.split("GPSNOISE1", 1)
        print(noise)
        print(main)
        
        # Construct the folder path for the log
        log_folder_path = os.path.join(log_folder, log_name1)

            
       


        with h5py.File(log_folder_path, "r") as file:
            # Assuming you have a dataset called 'frames' in the HDF5 file
            frames_dataset = file[main]["Image data"]
            num_frames = len(frames_dataset)
            print(num_frames)

                # Select a random frame index
            random_frame_index = random.randint(0, num_frames - 1)
            random_frame = frames_dataset[random_frame_index]

                # Display the random frame
            plt.imshow(random_frame, cmap='viridis')  # You can change the colormap as needed
            plt.title(f"Random Frame from {log_file}")
            plt.axis('off')
            plt.show()


if __name__ == "__main__":
    log_folder = r"C:\Users\bjqb7h\Downloads\Thesis2022\NUMPY\New\New Logs\tr\Bad logs"

    print("Visualizing GPS logs:")
    visualize_random_frame(log_folder)
```
